```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"G:\EDA\titanic.csv")
df.head()
```

```python
df.info()
```

```python
df['Age'].replace(np.nan , df['Age'].mean(), inplace=True)
```

```python
df['Fare'].replace(np.nan , df['Fare'].mean(), inplace=True)
```

```python
df.drop(['Cabin'] , axis =1 , inplace = True)
df.info()
```

```python
# Exploratory data analysis
Embarked_counts = df['Embarked'].value_counts()
Embarked_counts_df = Embarked_counts.to_frame().reset_index()
Embarked_counts_df.columns = ['Embarked', 'count']
print(Embarked_counts_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of passenger ages
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Passenger Ages')
plt.show()
```

```python
# Survival rate by Pclass
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Pclass')
plt.show()
```

```python
# Correlation matrix

numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numeric_columns].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

```

```python
# Survival rate by Sex
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.show()
```
